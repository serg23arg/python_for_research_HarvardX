```python
from math import sqrt

x = (0,0)
y = (1,1)

def distance(x,y):
   # define your function here!
  distance = sqrt((x[0]-y[0])**2 + (x[1]-y[1])**2)
  return distance

distance(x,y)
```

### Exercise 3

A list of numbers representing measurements obtained from a system of interest can often be noisy. One way to deal with noise to smoothen the values is by replacing each value with the average of the value and the values of its neighbors.

#### Exercise 3a

- Write a function `moving_window_average(x, n_neighbors)` that takes a list `x` and the number of neighbors `n_neighbors` on either side of a given member of the list to consider.
- For each value in `x`, `moving_window_average(x, n_neighbors)` computes the average of the value and the values of its neighbors.
- `moving_window_average` should return a list of averaged values that is the same length as the original list.
- If there are not enough neighbors (for cases near the edge), substitute the original value for a neighbor for each missing neighbor.
- Use your function to find the moving window sum of `x=[0,10,5,3,1,5]` and `n_neighbors=1`.

```python
def moving_window_average(x, n_neighbors=1):
    n = len(x)  # 6
    width = n_neighbors*2 + 1 # Esta variable contiene el ancho de la ventana
    x = [x[0]]*n_neighbors + x + [x[-1]]*n_neighbors #Esto agrega los valores de substituto al principio y al final de la lista, por concatenación.
      # To complete the function,
    # return a list of the mean of values from i to i+width for all values i from 0 to n-1.
    smooth_list = list()

    for i in range(n):
      avg = sum(x[i:i+width]) / width
      smooth_list.append(avg)

    return smooth_list

x = [0,10,5,3,1,5]

print(moving_window_average(x, 1))

```

```python
# write your code here!

avg_ranges = list()

for l in Y:

  max_value = l[0]
  min_value = l[0]

  for e in l:

    if e > max_value:
      max_value = e

    if e < min_value:
      min_value = e

  calc_range = max_value - min_value
  avg_ranges.append(calc_range)

print(avg_ranges)
```
